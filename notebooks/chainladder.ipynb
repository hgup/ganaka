```python
import chainladder as cl
import pandas as pd
```

### Data

```python
triangle = cl.load_sample("genins")
triangle
```

```python
latest = triangle.latest_diagonal
```

### Development

```python
cl_estimator = cl.Chainladder()

cl_model = cl_estimator.fit(triangle)

ultimates = cl_model.ultimate_

full_projection = cl_model.full_triangle_
```

```python
reserves = ultimates - triangle.latest_diagonal
reserves.to_frame()
```

```python
pd.concat([latest.to_frame(), ultimates.to_frame(), reserves.to_frame()], axis=1, keys=["Latest", "Ultimate", "Reserve"])
```
